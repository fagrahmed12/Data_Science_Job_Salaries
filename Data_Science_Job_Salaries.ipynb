```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import plotly.express as px
```

```python
df=pd.read_csv("/content/data science jobs.csv")
df
```

```python
df.info()
```

```python
#to check if there duplicated data or not
df.duplicated().sum()
```

```python
#to check if there null value or not
df.isnull().sum()
```

```python
top10_job= df['job_title'].value_counts()[:10]
plot = px.bar(y=top10_job.values, 
             x=top10_job.index,     
             color = top10_job.index,
             title= 'Top 10 Job Titles' )

plot.show()
```

```python
experience_level = df['experience_level'].value_counts()

plot = px.pie(df, values=experience_level.values, names=experience_level.index, title='Experience Level',
             color = experience_level.index)
plot.show()
```
